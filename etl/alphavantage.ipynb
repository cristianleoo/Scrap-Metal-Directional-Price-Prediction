```python
from pprint import pprint
import pandas as pd
```

```python
import requests

# Assuming you have a formatted_date variable
# time_from={formatted_date}T0130&

url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&topic=financial_markets&time_from=20210101T0130&apikey=8OA18HMQ8KC62Q10'
r = requests.get(url)
data = r.json()
data
```

```python
# Initialize a list to store the extracted information
extracted_data = []

# Loop through each entry in the 'feed'
for entry in data['feed']:
    # Check if 'ticker_sentiment' is not empty
    if entry['ticker_sentiment']:
        # Extract the desired information
        extracted_info = {
            'time_published': entry['time_published'],
            'relevance_score': entry['ticker_sentiment'][0]['relevance_score'],
            'ticker': entry['ticker_sentiment'][0]['ticker'],
            'ticker_sentiment_label': entry['ticker_sentiment'][0]['ticker_sentiment_label'],
            'ticker_sentiment_score': entry['ticker_sentiment'][0]['ticker_sentiment_score']
        }
    else:
        # If 'ticker_sentiment' is empty, set all fields to None
        extracted_info = {
            'time_published': None,
            'relevance_score': None,
            'ticker': None,
            'ticker_sentiment_label': None,
            'ticker_sentiment_score': None
        }

    # Append the extracted information to the list
    extracted_data.append(extracted_info)

# Create a DataFrame from the extracted data
df = pd.DataFrame(extracted_data)
df.dropna(how='all', inplace=True)
# Print the DataFrame
print(df)
```

```python
from datetime import datetime

# Assuming you have a date_string variable
date_string = '20231010T180021'

# Convert the string to a datetime object
def convert_date(date_string):
    try:
        #keep everything before the T
        date_string = date_string.split('T')[0]
        dt = datetime.strptime(date_string, '%Y%m%d')
        dt = pd.to_datetime(dt)
        return dt
    except ValueError:
        return date_string

convert_date(date_string)
```
