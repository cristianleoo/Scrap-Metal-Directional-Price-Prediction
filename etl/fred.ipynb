## FERROUS METALS (COMPLETE TABLE)

```python
df_ferrous_newor.head()
```

```python
ferrous_metal_data = pd.merge(df_ferrous_price_m[['date', 'value']], df_ferrous_inv[['date', 'value']], on='date', how='left', suffixes=('_price', '_inv'))
ferrous_metal_data
```

```python
ferrous_metal_data = pd.merge(ferrous_metal_data, df_ferrous_newor[['date', 'value']], on='date', how='left', suffixes=('', '_newor'))
ferrous_metal_data
```

```python
ferrous_metal_data = ferrous_metal_data.rename(columns = {'value_x': 'producer_price_index',
                           'value_y': 'total_inventory',
                           'value': 'new_orders'})
```

```python
ferrous_metal_data = ferrous_metal_data.drop(columns=['realtime_start_x', 'realtime_end_x', 'realtime_start_y','realtime_end_y', 'realtime_start', 'realtime_end'])
```
